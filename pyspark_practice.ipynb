```python
df.select(df.Pets,df.Education,
          when(df.Pets==0,"nopets") \
         .when(df.Pets<=3,"lessPets") \
         .when(df.Pets>3, "morePets") \
         .otherwise(df.Pets).alias("new_Pets") \
    ).show(5)

df.withColumn("new_pets",
          when(df.Pets==0,"nopets") \
         .when(df.Pets<=3,"lessPets") \
         .when(df.Pets>3, "morePets") \
         .otherwise(df.Pets)
    ).show(5)
```

def convertCase(str):
    resStr=""
    arr = str.split(" ")
    for x in arr:
       resStr= resStr + x[0:1].upper() + x[1:len(x)] + " "
    return resStr
